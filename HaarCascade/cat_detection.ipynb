```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 

def imshow(title ="image",image = None, size= 10 ):
    w, h =image.shape[0],image.shape[1]
    aspect_ratio =  w/h
    plt.figure(figsize=(size*aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
cat_classifier = cv2.CascadeClassifier(r'C:\Users\cengh\Desktop\ComputerVison\HaarCascade\haarcascade_frontalcatface.xml')
image = cv2.imread(r'C:\Users\cengh\Desktop\ComputerVison\HaarCascade\cat.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cat_face = cat_classifier.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
# Yüz bulunamadığında mesaj yazdır
if len(cat_face) == 0:
    print("No faces found")

# Yüzlerin üzerine dikdörtgen çiz
for (x, y, w, h) in cat_face:
    cv2.rectangle(image, (x, y), (x+w, y+h), (127, 0, 255), 2)

# Görüntüyü göster
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Face Detection')
plt.axis('off')
plt.show()
```
