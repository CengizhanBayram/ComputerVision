```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt

```

```python
# Define our imshow function 
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread(r'C:\Users\cengh\Desktop\ComputerVison\opencvbasics\2.jpg')
imshow("input message",image)
```

```python
image=cv2.imread(r'C:\Users\cengh\Desktop\ComputerVison\opencvbasics\2.jpg',0)
imshow("input message",image)
#Binary threshold
ret ,thresh1 = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
imshow("threshol bin0",thresh1)

blur =cv2.GaussianBlur(image,(5,5),0)
_,thresh3 =cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
imshow('gaussian otsu ', thresh3)
```

Thresholding, görüntü işleme ve bilgisayarla görme (computer vision) alanında çok önemli bir tekniktir. Bu teknik, bir görüntüyü belirli bir eşik değeri kullanarak basitleştirme amacı taşır ve çeşitli uygulama alanlarında birçok avantaj sağlar. İşte thresholding'in temel kullanım alanları ve sağladığı faydalar:

1. Nesne Tespiti (Object Detection)
Thresholding, bir görüntüdeki nesneleri arka plandan ayırmak için kullanılır. Örneğin, bir sahnedeki parlak nesneleri tespit etmek için, eşik değeri kullanılarak nesneler beyaz, arka plan ise siyah yapılabilir. Bu sayede, nesnelerin pozisyonları ve şekilleri daha kolay tespit edilir.

Kullanım Örneği: Trafik kameralarındaki araçları tespit etmek, endüstriyel robotlar için parça tanıma.

2. Segmentasyon (Segmentation)
Görüntü segmentasyonu, görüntüyü anlamlı bölgelere ayırma sürecidir. Thresholding, bu bölgelerin sınırlarını belirlemek ve her bölgeyi ayırt etmek için kullanılabilir. Örneğin, tıbbi görüntülerde farklı dokuları veya hücreleri ayırmak için kullanılabilir.

Kullanım Örneği: Tıbbi görüntülerde tümör veya lezyon tespiti, uydu görüntülerinde tarım arazilerinin ayrıştırılması.

3. Kenarlık Tespiti (Edge Detection)
Thresholding, kenarlık tespiti işlemlerinde de kullanılır. Görüntüdeki keskin yoğunluk değişimlerini belirlemek için eşik değeri uygulanır. Bu, görüntüdeki sınırları ve kenarları belirlemeye yardımcı olur.

Kullanım Örneği: Yüz tanıma sistemlerinde yüz hatlarının tespiti, haritalarda yol kenarlarının belirlenmesi.

4. Karakter Tanıma (Character Recognition)
Optik karakter tanıma (OCR) sistemlerinde, metinleri görüntülerden çıkarmak için thresholding kullanılır. Bu teknik, metin karakterlerini arka plandan ayırarak harflerin ve rakamların tanınmasını kolaylaştırır.

Kullanım Örneği: Doküman tarama ve dijitalleştirme, plakalı tanıma sistemleri.

5. Gürültü Giderme (Noise Reduction)
Thresholding, belirli bir yoğunluğun altındaki veya üstündeki piksel değerlerini sıfırlayarak görüntülerdeki gürültüyü azaltabilir. Bu, özellikle düşük ışık koşullarında çekilmiş görüntülerde yararlıdır.

Kullanım Örneği: Güvenlik kamerası görüntülerindeki gürültünün azaltılması, astronomik görüntülerin iyileştirilmesi.

6. Hareket Tespiti (Motion Detection)
Hareket tespiti sistemlerinde, ardışık video kareleri arasındaki farkları tespit etmek için thresholding kullanılır. Bu sayede, sahnede hareket eden nesneler belirlenir.

Kullanım Örneği: Güvenlik sistemlerinde hareketli nesnelerin tespiti, spor analizinde oyuncu hareketlerinin izlenmesi.

7. Renk Segmentasyonu (Color Segmentation)
Thresholding, belirli bir renk aralığındaki pikselleri ayırmak için kullanılabilir. Bu, renk tabanlı nesne tespiti ve izleme uygulamalarında faydalıdır.

Kullanım Örneği: Otonom araçlarda trafik ışığı ve yol işareti tespiti, tarım uygulamalarında bitki hastalıklarının izlenmesi.

Thresholding'in Faydaları
Basit ve Hızlı: Hesaplama açısından oldukça verimlidir ve gerçek zamanlı uygulamalar için uygundur.
Kolay Uygulanabilir: Temel bir işlem olduğundan, birçok farklı uygulamada rahatlıkla kullanılabilir.
Görüntü İyileştirme: Görüntüdeki önemli özellikleri vurgulayarak daha iyi analiz edilmesini sağlar.
