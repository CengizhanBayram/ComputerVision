```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt

```

```python
# Define our imshow function 
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread(r'C:\Users\cengh\Desktop\ComputerVison\opencvbasics\2.jpg')
imshow("input message",image)
```

```python
image=cv2.imread(r'C:\Users\cengh\Desktop\ComputerVison\opencvbasics\2.jpg',0)
imshow("input message",image)
#Binary threshold
ret ,thresh1 = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
imshow("threshol bin0",thresh1)

blur =cv2.GaussianBlur(image,(5,5),0)
_,thresh3 =cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
imshow('gaussian otsu ', thresh3)
```
